```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_final = pd.read_csv("./Class#12_Pandas_Python/car_final.csv")
car_insurance = pd.read_csv("./Class#12_Pandas_Python/car_insurance.csv")
```

```python
car_final.head()
```

```python
car_insurance.head()
```

```python
merged_data = car_final.merge(car_insurance)
merged_data
```

```python
merged_data.sort_values(["car"])
```

```python
moderate = merged_data[merged_data.insurance == "moderate"]
moderate_count = moderate.groupby(["car"]).count()
moderate_count
```

```python
moderate_count.plot(kind="bar")
plt.show()
```

```python
merged_data[["maintenance"]].plot(
    kind="hist",
    bins=[1000, 1500, 2000, 2500, 3000, 3500, 4000, 5000, 5500, 6000],
)
plt.show()
```
