```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
zoo_final = pd.read_csv("./Class#12_Pandas_Python/zoo_final.csv")
zoo_weight = pd.read_csv("./Class#12_Pandas_Python/zoo_weight.csv")
```

```python
zoo_final.head()
```

```python
zoo_weight.head()
```

```python
merged_data = zoo_final.merge(zoo_weight, how="left").fillna("unknown")
merged_data
```

```python
moderate = merged_data[merged_data.weight == "moderate"]
moderate_count = moderate.groupby(["animal"]).count()
moderate_count
```

```python
moderate_count.plot(
    kind="bar", title="Animal by Weight", xlabel="Animal Type", ylabel="Amount"
)
plt.show()
```
